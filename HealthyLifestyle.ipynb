```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv into dataframe
df = pd.read_csv('/content/healthy_lifestyle_city_2021.csv')
```

```python
print('Data shape: ', df.shape)
print('-----------------------------')
df.head()
```

```python
# check if any values are null or contain missing values
df.isnull().sum()
```

```python
for x in df.columns:
  uniq = np.unique(df[x])
  values = len(uniq)
  if values < 12:
      print('The # of values for {} :{} -- {}'.format(x, values, uniq))
  else:
      print('The # of values for {} :{}'.format(x, values))
```

```python
df.describe()
```

```python
def plot_hist(features, dataframes):
  for f in features:
    plt.figure(figsize=(10,6))
    ax = sns.displot(x=f, data=dataframes, palette="mako")
```

```python
df.columns
```

```python
df.dtypes
```

```python
feats = df[['Rank', 'Life expectancy(years) (Country)', 'Annual avg. hours worked', 
            'Outdoor activities(City)', 'Number of take out places(City)']]
```

```python
plot_hist(feats, df)
```

```python
df.columns
```

```python
def plot_count(features, dataframes):
  for f in features:
    plt.figure(figsize=(10,6))
    ax = sns.countplot(x=f, data=dataframes, palette="mako")
    plt.xticks(rotation=45)
```

```python
feats_2 = df[['Sunshine hours(City)','Cost of a bottle of water(City)', 
              'Obesity levels(Country)', 'Annual avg. hours worked',
              'Sunshine hours(City)']]
```

```python
plot_count(feats_2, df)
```

```python
def barplot_count(features, dataframes, c):
  for f in features:
    plt.figure(figsize=(14,6))
    ax = sns.barplot(x=f, y=c, data=dataframes, palette="mako")
    plt.xticks(rotation=90)
```

```python

def boxplot_count(features, dataframes):
  for f in features:
    plt.figure(figsize=(10,6))
    ax = sns.boxplot(x=f, data=dataframes, palette="mako")
```

```python
feats_4 = df[['Rank', 'Life expectancy(years) (Country)',  
            'Outdoor activities(City)', 'Number of take out places(City)']]
boxplot_count(feats_4, df)
```
